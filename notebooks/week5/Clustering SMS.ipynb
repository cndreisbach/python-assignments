```python
import csv
import pandas as pd
from textblob import TextBlob
from sklearn.cluster import KMeans
```

```python
sms = pd.read_csv("SMSSpamCollection", sep="\t", names=['class', 'text'])
sms = sms.drop("class", axis=1)
```

```python
sms.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=500)
corpus = cv.fit_transform(sms["text"])
```

```python
corpus.toarray()[0].nonzero()
```

```python
cv.get_feature_names()
```

```python
len(cv.get_feature_names())
```

```python
kmeans = KMeans(4)
```

```python
results = kmeans.fit_predict(corpus)
```

```python
sms['class'] = results
```
